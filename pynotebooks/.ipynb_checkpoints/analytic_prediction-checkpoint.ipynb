```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams.update({'font.size': 16})
```

```python
def pot_g(r, R):
    if r < R:
        return M*(3*R**2-r**2)/(2*R**3)
    else:
        return M/r
    
def transform_chi(chi):
    return psi_s*(1-chi)
    
def chi_B(R, delta=None):
    if delta is None:
        delta = M/(4/3.*np.pi*R**3)
    chi = 1/(1+delta)**2
    return transform_chi(chi)

def chi_sol(r, R, delta=None):
    m_scr = np.exp(-mass*r)
    if r <= R:
        if pot_g(r, R) < psi_s:
            return m_scr*pot_g(r, R)
        else:
            return m_scr*chi_B(R, delta)
    else:
        screen = chi_sol(R, R) / pot_g(R, R)
        if screen > 1:
            screen = 1
        return pot_g(r, R)*screen*m_scr
    
def plot_pot_g(ax, R, c='g', ls='-'):
    r = np.logspace(-1, 3)
    phi = np.array([pot_g(x, R) for x in r])
    ax.plot(r, phi, c=c, ls=ls, label='$\Phi_G (r, R = %.1f)$' % R)
    ax.plot(R, pot_g(R, R), 'k+')
    
def plot_psi_s(ax, c='r', ls='-'):    
    ax.axhline(y=psi_s, c=c, ls=ls, label='$\Phi_s=%.1f$' % psi_s)
    
def plot_chi_B(ax, R, c='r', ls='--'):
    chi = chi_B(R)
    ax.axhline(y=chi, c=c, ls=ls, label='$\chi_B (R = %.1f) = %.2f$' % (R, chi))
    
def plot_chi(ax, R, c='b', ls='-'):
    r = np.logspace(-1, 3)
    chi = np.array([chi_sol(x, R) for x in r])
    ax.plot(r, chi, c=c, ls=ls, label='$\chi (r, R = %.1f)$' % R)
    ax.plot(R, chi_sol(R, R), 'k+')
    
```

```python
M = 20
psi_s = 10.0
O_m = 0.3089
H0 = 67.74
h = H0 / 100
c_kms = 299792.458
n = 0.2
L = 100
chi_pref = 3/2.*O_m*pow(L * H0 * h / c_kms ,2) / psi_s
mass_sq = (1-n) * chi_pref
mass = np.sqrt(mass_sq)
```

```python
fig = plt.figure(figsize=(12, 8), dpi=120)
ax = plt.gca()

all_R = [1, 8.]
all_ls = ['--', '-.', ':']

# screening potential
plot_psi_s(ax)

for R, ls in zip(all_R, all_ls):
    # bulk field
    plot_chi_B(ax, R, ls=ls)
    
    # gravitational potential
    plot_pot_g(ax, R, ls=ls)
    
    # chameleon filed
    plot_chi(ax, R, ls=ls)

# r axis
plt.xscale('log')
plt.xlabel(r'r')
    
# phi axis
ax.set_yscale('symlog', linthreshy=1e-2)
ax.set_ylabel(r'$\Phi_G$')
ax.axhline(y=0, c='k', ls='--')

# chi axis
ax2 = ax.twinx()
ax2.set_yscale('symlog', linthreshy=1e-2)
ax2.set_ylim(*ax.get_ylim())
yticks = [psi_s, 0] # , chi_B(8)
yticklabels = ['0', r'$\chi_0$'] # , r'$\chi_B(R = 8)$'
ax2.set_yticks(yticks)
ax2.set_yticklabels(yticklabels)
ax2.set_ylabel(r'$\chi$')

# legend
ax.legend()

# save
out_dir='/home/vrastil/Documents/GIT/Adhesion-Approximation/output/test_runs/test_ChiSolver/'
fig.savefig(out_dir + 'anal_chi/chi.png')
```
